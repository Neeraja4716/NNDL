```
x_o=2
eta=0.01
eps=0.000001
del_x=1
max_iters=100
iters=0
def deriv(x):
  x_deriv=4*((x)**3)+6*(x)
  return x_deriv
while abs(del_x)>eps and iters < max_iters:
  prev_x=x_o
  del_x=-eta*deriv(prev_x)
  x_o=x_o+del_x
  iters=iters+1
  print("iteration",iters,"\n x value is",x_o)
print("the local minimum occurs at",x_o)
```
